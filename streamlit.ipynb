```python
# json to dataframe
import pandas as pd
import json

def dataframe(filename):
    with open(filename, 'r') as f:
        data = json.load(f)

    rows = []
    for result in data['results']:
        location = result['location']
        description = location.get('description', None) # use the .get() method to avoid KeyError
        length = location.get('length', None) 
        links = location['shape']['links']
        latitudes = []
        longitudes = []
        for link in links:
            if 'points' in link:
                for point in link['points']:
                    latitudes.append(point['lat'])
                    longitudes.append(point['lng'])
        currentFlow = result.get('currentFlow', {})
        speed = currentFlow.get('speed', None) 
        freeflow = currentFlow.get('freeFlow', None) 
        jamfactor = currentFlow.get('jamFactor', None) 
        rows.append([description, length, latitudes, longitudes, speed, freeflow, jamfactor])

    return pd.DataFrame(rows, columns=['Description', 'Length', 'Latitudes', 'Longitudes', 'Speed', 'FreeFlow', 'JamFactor'])
```

```python
#processing files 
import glob
import os
import pandas as pd
import numpy as np
import json
import pickle
import datetime
def process_here_maps_outputs(directory_path):
    # Get a list of all JSON files in the directory
    json_files = glob.glob(os.path.join(directory_path, '*.json'))

    # Define empty DataFrame before the loop
    df = pd.DataFrame(columns=['Time', 'Date', 'Day', 'Description', 'Length', 'Latitudes', 'Longitudes', 'Speed', 'FreeFlow', 'JamFactor'])
    error_count = 0
    error_files = []
    
    for json_file in json_files:
        # Get the filename and parse the date and time
        filename = os.path.basename(json_file)
        date, time = os.path.splitext(filename)[0].split()
        year, month, day = date.split('-')
        hour, minute, second = time.split('-')
        
        # Get the day of the week as a string
        day_of_week = datetime.datetime(int(year), int(month), int(day)).strftime('%A')
        
        # Process each JSON file using the existing logic
        try: 
            df_temp = dataframe(json_file)
            print(f'{filename} ok dataframe')
        except:
            print(f'{filename} error dataframe')
            error_count += 1
            error_files.append(filename)
            continue
        
        # Add the date, time, and day columns to the DataFrame
        df_temp['Date'] = date
        df_temp['Time'] = time
        df_temp['Day'] = day_of_week
        
        # Append to the DataFrame after processing each file
        df = pd.concat([df, df_temp])
    
    print(f'Total number of error files: {error_count}')  
    print("Files with errors:")
    print(error_files)
    return df
```

```python
import time
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="my_app")

def search_kecamatan(lat, lon):
    location = None
    while not location:
        try:
            location = geolocator.reverse(f"{lat}, {lon}")
        except Exception as e:
            print(f"Error: {str(e)}. Retrying in 1 second...")
            time.sleep(3)
    try:
        kecamatan = location.raw['address']['subdistrict']
        print('Kecamatan found:', kecamatan)
        return kecamatan
    except KeyError:
        print('Error: Kecamatan not found')
        return None

# apply the function to the dataframe
df_monday_unique['Kecamatan'] = df_monday_unique.apply(lambda x: search_kecamatan(x['Latitudes'][0], x['Longitudes'][0]), axis=1)
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import time

# load the data
df_00 = pd.read_csv('traffic_bandung_00_unique.csv')

# initialize the geolocator and the rate limiter
geolocator = Nominatim(user_agent="geoapiExercises")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

# create a function to get the location of a kecamatan using geopy
def get_location_kecamatan(kecamatan):
    try:
        location = geocode(f"{kecamatan}, Bandung, Indonesia")
        if location is None:
            print(f"No location found for {kecamatan}")
            return None
        else:
            print(f"{kecamatan} is at {location.latitude}, {location.longitude}")
            return location.latitude, location.longitude
    except Exception as e:
        print(f"Error while geocoding {kecamatan}: {str(e)}")
        return None

# apply the function to get the location of each kecamatan
df_00['location_kecamatan'] = df_00['Kecamatan'].apply(lambda x: get_location_kecamatan(x))
time.sleep(5)  # wait for 5 seconds before proceeding to the next request

# split the location column into two columns: latitude and longitude
df_00[['latitude_kecamatan', 'longitude_kecamatan']] = pd.DataFrame(df_00['location_kecamatan'].tolist(), index=df_00.index)

# save the new dataframe
df_00.to_csv('new_file.csv', index=False)
```
